```python
from minbc_lib.dwwallets import *
wallet1 = get_wallet(0)

y = []
for i in wallet1.query(class2='電気'):
    y = y + [-i[0]]
```

```python
from matplotlib import pyplot as plt
import numpy as np

# 0~11までのリストを作る
x = np.arange(12)

# グラフの範囲を決める
plt.ylim([0, 5500])
# 描画
plt.plot(x, y)
```

```python
print(y)
```

```python
# np.polyfitを使って近似関数を求める
# 今回は5次関数を求める。切片を含め、5次関数には6つのパラメータ（係数）が必要となる

coef = np.polyfit(x, y, 5)
```

```python
print(coef)
```

```python
x2 = np.arange(0, 11, 0.1)
func = np.poly1d(coef) # 係数から関数を得る
y2 = func(x2)

plt.ylim([0, 5500])
plt.plot(x, y)
plt.plot(x2, y2)
```
